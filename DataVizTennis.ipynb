```python
#ATP points evolution for all winners in percentage from the beginning of their season to the end

total_players_w = list(df_tennis['winner'].unique())

pct_evolution = {}

for col in total_players_w:
    #pct_evolution = {}
    list_ptn = list(df_tennis[df_tennis['winner'] == col]['wpts'])
    pct_evolution.update({col:round(((list_ptn[-1] - list_ptn[0])/(list_ptn[0])) * 100,2)}) 
    
print(pct_evolution)
```

```python
all_values = pct_evolution.values()
max_value = max(all_values)
print(max_value)
```

```python
def get_keys_from_value(d, val):
    return [k for k, v in d.items() if v == val]
```

```python
#player who gained the most points
keys = get_keys_from_value(pct_evolution, 202.88)
print(keys)
```
